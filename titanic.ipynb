```python
# Cleaning data
# I have alot of columns for zeros that i have no idea what to use, so let me firt drop them
new_df = titanic_df.drop(titanic_df.columns[[5, 6, 7, 8, 9, 10, 11, 13, 14, 15, 16, 17, 18, 19, 20, 22, 23, 25, 26]], axis=1)

# Also let's change the misspelled Survived
renamed_df = new_df.rename(columns={'2urvived': 'Survived'})
```

```python
renamed_df.to_csv('train_and_test2.csv', index=False)
renamed_df.head()
```

```python
names_df = pd.read_csv('names.csv')
names_df.head()
```

```python
merged_df = pd.merge(renamed_df, names_df, on='Passengerid', how='left')
merged_df.head()
```

```python
# Re-order the columns
# Get the current column order
columns = merged_df.columns.tolist()
```

```python
# Move the columns to the desired positions
# Temporarily remove the name column
columns.remove('Names')

# Insert the desired order
columns.insert(columns.index('Passengerid') + 1, 'Names')

# Re-order the DataFrame
merged_df = merged_df[columns]
merged_df.head()
```

```python
merged_df.to_csv('new_cleaned_dataset.csv', index=False)
```
